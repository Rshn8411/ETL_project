```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
file = "raw_data/State_population_data_2010-2019.xlsx"
```

```python
# Read our Data file for 2019 with the pandas library
population = pd.read_excel(file)
pop_df = pd.DataFrame(population)
```

```python
pop_df.rename(columns={"Unnamed: 0":"state"},inplace=True)
pop_df.dropna(how='any',inplace=True)
pop_df.head()
```

```python
#identify index of unneeded regions
us = pop_df[pop_df['state'] == "United States"].index
northeast = pop_df[pop_df['state'] == "Northeast"].index
midwest = pop_df[pop_df['state'] == "Midwest"].index
south = pop_df[pop_df['state'] == "South"].index
west = pop_df[pop_df['state'] == "West"].index
```

```python
#list of regions to drop
drop_list = [us,northeast,midwest,south,west]
```

```python
state_only_pop = pop_df
```

```python
#for loop to get rid of the larger regions listed above
for region in drop_list:
    state_only_pop.drop(region,inplace=True)
```

```python
#rename column headers to something more suitable for db upload
state_only_pop.rename(columns={'Census':'census','Estimates Base':'estimates_base'})
```

```python
state_only_pop.reset_index(inplace=True)
state_only_pop.head()
```

```python
#list of states to insert into DF
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY","PR"]
```

```python
#insert in DF
state_only_pop['state_abbr'] = states
```

```python
#Get only the columns we need and reorder
state_only_pop = state_only_pop.filter(['state','state_abbr','Census','Estimates Base',2015,2016,2017,2018])
```
