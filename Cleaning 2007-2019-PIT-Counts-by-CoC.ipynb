```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
file = "Resources/2007-2019-PIT-Counts-by-CoC.xlsx"
```

```python
# Read our Data file for 2019 with the pandas library
readFile2019 = pd.read_excel(file, '2019')
```

```python
#Drop null values to avoid errors
readFile2019.dropna(inplace=True)

#Split CoC number to isolate state
readFile2019[['State','Number']] = readFile2019['CoC Number'].str.split("-", expand=True)
readFile2019
```

```python
#Group by State and sum values
group2019 = readFile2019.groupby('State').sum()
group2019
```

```python
#Print all column names
for column in group2019.columns:
    print(column)
```

```python
#Select only overall values for 2019
modified2019 = group2019[['Overall Homeless, 2019',
'Overall Homeless - Under 18, 2019',
'Overall Homeless - Age 18 to 24, 2019',
'Overall Homeless - Over 24, 2019',
'Overall Homeless - Female, 2019',
'Overall Homeless - Male, 2019',
'Overall Homeless - Transgender, 2019',
'Overall Homeless - Gender Non-Conforming, 2019',
'Overall Homeless - Non-Hispanic/Non-Latino, 2019',
'Overall Homeless - Hispanic/Latino, 2019',
'Overall Homeless - White, 2019',
'Overall Homeless - Black or African American, 2019',
'Overall Homeless - Asian, 2019',
'Overall Homeless - American Indian or Alaska Native, 2019',
'Overall Homeless - Native Hawaiian or Other Pacific Islander, 2019',
'Overall Homeless - Multiple Races, 2019']]
modified2019
```

```python
#Rename columns
renamed2019 = modified2019.rename(columns={'Overall Homeless, 2019': '2019 Overall Homeless',
'Overall Homeless - Under 18, 2019': '2019 Homeless Under 18',
'Overall Homeless - Age 18 to 24, 2019': '2019 Homeless Age 18 to 24',
'Overall Homeless - Over 24, 2019': '2019 Homeless Over 24',
'Overall Homeless - Female, 2019': '2019 Homeless Female',
'Overall Homeless - Male, 2019': '2019 Homeless Male',
'Overall Homeless - Transgender, 2019': '2019 Homeless Transgender',
'Overall Homeless - Gender Non-Conforming, 2019': '2019 Homeless Non-Conforming',
'Overall Homeless - Non-Hispanic/Non-Latino, 2019': '2019 Homeless Non-Hispanic/Non-Latino',
'Overall Homeless - Hispanic/Latino, 2019': '2019 Homeless Hispanic/Latino',
'Overall Homeless - White, 2019': '2019 Homeless White',
'Overall Homeless - Black or African American, 2019': '2019 Homeless Black or African American',
'Overall Homeless - Asian, 2019': '2019 Homeless Asian',
'Overall Homeless - American Indian or Alaska Native, 2019': '2019 Homeless American Indian or Alaska Native',
'Overall Homeless - Native Hawaiian or Other Pacific Islander, 2019': '2019 Homeless Native Hawaiian or Other Pacific Islander',
'Overall Homeless - Multiple Races, 2019': '2019 Homeless Multiple Races'}) 
renamed2019
```

```python
#Drop American territorries
final2019 = renamed2019.drop(['GU','MP', 'VI'])
final2019
```
