```python
#dependcies
import pymongo
import os
import pandas as pd
import json
```

```python
#connection and local host
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# database name
db = client.etlDB

```

```python
# function(two parameters) to convert csv files to json, name the collection, and print results 
def insert_csv(csv_name,collection_name):
    collection = collection_name
    data = pd.read_csv(r'resources/'+csv_name+'.csv')
    data_json = json.loads(data.to_json(orient='records'))
    collection.insert_many(data_json)
    
    results = collection.find()
    for item in results:
        print(item)
```

```python
# inserting data from the CoC (continuum of care) award csv
insert_csv('award_amount_coc',db.coc_award)
```

```python
#insert the state award data from csv 
insert_csv('award_by_state',db.state_award)
```

```python
#import state population data from csv
insert_csv('state_population',db.state_pop)
```

```python
#import state population data from csv
insert_csv('2015-2018_homeless_data',db.homeless_data)
```

```python
# Example query 1; homeless data
myquery = {'State':'UT'}
state_query = db.homeless_data.find(myquery)
for info in state_query:
    print(info)
```

```python
# Example query 2; Population query for UT
pop_query = {'state_abbr':'UT'}
ca_query = db.state_pop.find(pop_query)
for pop_info in ca_query:
    print(pop_info)
```

```python
# Example query 3; State Award
award_query = {'year':2015}
query_2015 = db.state_award.find(award_query)
for award in query_2015:
    print(award)
```
