```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
cost_csv = pd.read_csv(r"resources/CoC_awards_state.csv")
```

```python
cost_df = pd.DataFrame(cost_csv)
cost_df.head()
```

```python
amount_state = cost_df.filter(['Year','State','Award Amount'])
```

```python
amount_state.head()
```

```python
group_state = amount_state.groupby(['Year','State'])
```

```python
group_state.head()
```

```python
group_state=group_state.sum()
```

```python
group_state = group_state.reset_index()
```

```python
group_state.to_csv(r"resources/award_by_state.csv")
```
