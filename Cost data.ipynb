```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
cost_csv = pd.read_csv(r"resources/CoC_awards_state.csv")
```

```python
cost_df = pd.DataFrame(cost_csv)
cost_df.head()
```

```python
cost_df.index.name = "state_cost_id"
cost_df.rename(columns={"Year":"year","State":"state","Award Amount":"award_amount",
                        "CoC Name":"coc_name"},inplace=True)
```

```python
guam = cost_df[cost_df['state'] == "GU"].index
virIslands = cost_df[cost_df['state'] == "VI"].index
```

```python
cost_df.drop(guam,inplace=True)
cost_df.drop(virIslands,inplace=True)
```

```python
amount_state = cost_df.filter(['year','state','award_amount'])
```
